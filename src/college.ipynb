```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define project phases with start times and durations (in months)
phases = [
    {"phase": "Phase 1: Data Acquisition & Clinical Partnerships", "start": 1, "duration": 3},
    {"phase": "Phase 2: Hardware Prototyping & Model Development", "start": 4, "duration": 1.5},
    {"phase": "Phase 3: Dataset Validation & Field Testing", "start": 6, "duration": 1},
    {"phase": "Phase 4: Deployment & Refinement", "start": 6.5, "duration": 0.5},
    {"phase": "Phase 5: Reporting & Scalability", "start": 8, "duration": 1}
]

# Define distinct colors for each phase
colors = ["skyblue", "lightgreen", "salmon", "orchid", "gold"]

fig, ax = plt.subplots(figsize=(20, 6))

# Create horizontal bars for each phase and collect handles for legend
handles = []
for i, phase in enumerate(phases):
    ax.barh(i, phase["duration"], left=phase["start"], height=0.4, color=colors[i], edgecolor='black')
    # Center the phase label in the bar
    # ax.text(phase["start"] + phase["duration"]/2, i, phase["phase"],
    #         va='center', ha='center', color='black', fontsize=8, wrap=True)
    # Create a patch for the legend
    patch = mpatches.Patch(color=colors[i], label=phase["phase"])
    handles.append(patch)

# Configure the chart
ax.set_yticks([])
ax.set_xlabel("Month")
ax.set_title("NIROGYA Project Timeline")
ax.set_xlim(0, 10)
ax.set_xticks(range(0, 11))
plt.tight_layout()

# Add legend at the bottom center with two columns
plt.legend(handles=handles, loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2)

plt.show()

```

```python
import matplotlib.pyplot as plt

# Define project phases with revised start times and durations (in months)
phases = [
    {"phase": "Phase 1: Data Acquisition & Clinical Partnerships", "start": 1, "duration": 3},
    {"phase": "Phase 2A: Hardware Prototyping", "start": 3.5, "duration": 1.5},
    {"phase": "Phase 2B: Model Development", "start": 5, "duration": 1},
    {"phase": "Phase 3A: Dataset Validation", "start": 6, "duration": 1},
    {"phase": "Phase 3B: Field Testing", "start": 7, "duration": 1},
    {"phase": "Phase 4: Deployment & Refinement", "start": 8, "duration": 0.5},
    {"phase": "Phase 5: Reporting & Scalability", "start": 8.5, "duration": 1}
]

# Define distinct colors for each phase
colors = ["skyblue", "lightgreen", "salmon", "orchid", "gold", "violet", "cyan"]

fig, ax = plt.subplots(figsize=(20, 6))

y_positions = range(len(phases))
phase_names = [phase["phase"] for phase in phases]

# Create horizontal bars for each phase
for i, phase in enumerate(phases):
    ax.barh(i, phase["duration"], left=phase["start"], height=0.4, color=colors[i], edgecolor='black')

# Set y-axis labels to the phase descriptions
ax.set_yticks(y_positions)
ax.set_yticklabels(phase_names)

# Configure the chart
ax.set_xlabel("Month")
ax.set_title("NIROGYA Project Timeline")
ax.set_xlim(0, 10)
ax.set_xticks(range(0, 11))
plt.tight_layout()

plt.show()
```
